```python
df['total_points'].describe()
```

```python
df.groupby('type_1')['total_points'].mean().sort_values(ascending=False)
```

Dragon type pokemon stand out as having more total stat points than most other types on average. Unsurprisingly, bug types have the lowest on average.

```python
df.groupby('type_1')['is_legendary'].mean().sort_values(ascending=False)
```

Dragon, psychic, and fairy have the highest percentage of legendary pokemon by far compared to the other types. This is only counting the type 1 however.

```python
df.groupby('type_2')['is_legendary'].mean().sort_values(ascending=False)
```

Other types such as ice, fire, and steel now have legendaries accounted for because they are more primarily in type 2.

```python
df.groupby('type_1')['speed'].mean().sort_values(ascending=False)
```

Flying, electric, and dragon types typically lead in speed.

```python
df.groupby('generation')['total_points'].mean().sort_values(ascending=False)
```

In general, more recent generations have had pokemon with higher total points on average.
